```python
###---Box-plot--
df = pd.DataFrame(df)

# Plot box plots for numeric variables with colors
numeric_vars = ['FFMC', 'DMC', 'DC', 'ISI', 'temp', 'RH', 'wind', 'rain', 'area']
boxplot = df[numeric_vars].boxplot(figsize=(15, 10), patch_artist=True, boxprops=dict(facecolor='blue'))
plt.title('Box Plot of Numeric Variables')
plt.show()
```

```python
##---Outliers--
numeric_vars = ['FFMC', 'DMC', 'DC', 'ISI', 'temp', 'RH', 'wind', 'rain', 'area']
boxplot = df[numeric_vars].boxplot(figsize=(15, 10), patch_artist=True, boxprops=dict(facecolor='red'))

# Calculate and mark outliers
for i, var in enumerate(numeric_vars):
    q1 = df[var].quantile(0.25)
    q3 = df[var].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr

    outliers = df[(df[var] < lower_bound) | (df[var] > upper_bound)][var]
    for value in outliers:
        plt.scatter(i + 1, value, color='red', marker='o')

plt.title('Box Plot of Numeric Variables with Outliers')
plt.show()
```

```python
##---Displot---
df = pd.DataFrame(df)
colors = ['blue', 'green', 'coral', 'yellow', 'pink', 'lightgray', 'orange', 'cyan', 'purple']
numeric_vars = ['FFMC', 'DMC', 'DC', 'ISI', 'temp', 'RH', 'wind', 'rain', 'area']
for i, var in enumerate(numeric_vars):
    plt.figure(figsize=(8, 6))
    sns.distplot(df[var], hist=True, kde=True, bins=20, color=colors[i], hist_kws={'edgecolor': 'black'}, kde_kws={'linewidth': 2})
    plt.title(f'Distribution Plot of {var}')
    plt.xlabel(var)
    plt.ylabel('Density')
    plt.show()
```

# Label encoding and transformation

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['month']=le.fit_transform(df['month'])
df['day']=le.fit_transform(df['day'])
df['size_category']=le.fit_transform(df['size_category'])
```

```python
x=df.iloc[:,0:30]
```

```python
y=df.iloc[:,30]
```

# importing Tensorflow , keras model and layers

```python
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
```

# input and output

```python
model=Sequential()
model.add(Dense(45,input_dim=30,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

# Binary cross_entropy and optimizers

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
history = model.fit(x, y, validation_split=0.30, epochs=500, batch_size=200)
```

# Accuracy results

```python
scores=model.evaluate(x,y)
```
